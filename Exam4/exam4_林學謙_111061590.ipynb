###1.1(10%): Build the environment

In this grid world, the tiles in green represent swamp. The yellow one represents the terminal state, and the blue one represents the initial state.

Given a state and an action, please write a function that returns the  next-state and the corresponding reward.

The input state is a list, and the action should be chosen from the variable 'ACTIONS'.

```python
def step1(state, action):
  ###### Your code here ######
  SWAMP = [[0,2],[0,3],[0,4],[0,5],[2,2],[2,3],[2,4],[2,5],[3,2],[3,3],[4,2],[4,3],[4,4]]
  s1 = np.array(state) + np.array(action)
  key = 0
  if s1[0] >= WORLD_SIZE or s1[1] >= WORLD_SIZE or s1[0] <= -1 or s1[1] <= -1:
    s1 = state
    key = 1

  if key == 0:
    state = s1

  if rd.rand() < 0.3:
    s1 = np.array(s1) - np.array([0,1])
  elif rd.rand() < 0.3:
    s1 = np.array(s1) + np.array([0,1])

  if s1[0] >= WORLD_SIZE or s1[1] >= WORLD_SIZE or s1[0] <= -1 or s1[1] <= -1:
    s1 = state

  r1 = -1
  #for check swamp
  s1 = list(s1)
  if s1 in SWAMP:
    r1 = -100
  ################################
  return [s1,r1]
```

###1.2(10%): Sarsa

Using the environment in 1.1, please implement Sarsa with $\gamma=1 $, $\alpha=0.1$ and $\epsilon=0.1$ over 500 episodes.

The output 'returns' is a list consisting of return 'G' from each episode.

```python
GAMMA=1.0
ALPHA=0.1
EPSILON=0.1
```

```python
def chooseAction(state, qtable, epsilon):
  if rd.rand()>=epsilon:
    return np.argmax(qtable[state[0],state[1]])
  else:
    return rd.randint(4)
```

```python
def sarsa():
  ###### Your code here ######
  qtable =rd.rand(6,6,4)
  qtable[1][5]=0
  returns = []
  for episode in range(500):
    state=S0
    total_reward=0
    action=chooseAction(state, qtable, EPSILON)
    while state != ST:
      next_state,reward=step1(state,ACTIONS[action])
      next_action=chooseAction(next_state, qtable, EPSILON)
      qtable[state[0]][state[1]][action] += ALPHA*(reward + GAMMA*qtable[next_state[0]][next_state[1]][next_action] - qtable[state[0]][state[1]][action])   
      state=next_state
      action=next_action
      total_reward += reward     
    returns.append(total_reward)
  ################################
  return returns
```

```python
test=moving_averages(np.average([sarsa() for i in range(5)],axis=0))
plt.plot(test,label='sarsa()')
plt.ylim(-500, 0)
plt.legend()
plt.show()
```

###1.3(20%): Sarsa(λ)

Using the environment in 1.1, please implement Sarsa(λ) with $\gamma=1 $, $\alpha=0.1$, $\epsilon=0.1$ and  $λ=0.5$ over 500 episodes.

The output 'returns' is a list consisting of return 'G' from each episode.

```python
GAMMA=1.0
ALPHA=0.1
EPSILON=0.1
LAMBDA=0.5
```

```python
def sarsa_L():
  ###### Your code here ######
  qtable = rd.rand(6,6,4)
  qtable[1][5]=0
  returns = []

  for episode in range(500):
    state=S0
    total_reward=0
    action=chooseAction(state, qtable, EPSILON)
    ztable=np.zeros([6,6,4])
    while state != ST:
      next_state, reward = step1(state,ACTIONS[action])
      next_action = chooseAction(next_state, qtable, EPSILON)
      delta = reward + GAMMA * qtable[next_state[0]][next_state[1]][next_action] - qtable[state[0]][state[1]][action]     
      ztable[state[0]][state[1]][action] += 1            
      qtable += delta*ALPHA*ztable
      ztable = GAMMA*LAMBDA*ztable
      state=next_state
      action=next_action
      total_reward += reward
    returns.append(total_reward)
  ################################
  return returns
```

```python
test=moving_averages(np.average([sarsa_L() for i in range(5)],axis=0))
plt.plot(test,label='sarsa_L()')
plt.ylim(-500, 0)
plt.legend()
plt.show()
```

### 1.4(20%): Actor-Critic

Using the environment in 1.1, please implement Actor-Critic with $\gamma=1 $,  learning rate for policy $\alpha_\theta=0.001$, learning rate for value state $\alpha_v=0.1$ over 500 episodes.

The output 'returns' is a list consisting of return 'G' from each episode.

```python
GAMMA = 1.0
ALPHA_theta = 0.001
ALPHA_v = 0.1
```

```python
# use policy choose action 5%
def chooseAction_pg(pi):
  ###### Your code here ######
  action = np.random.choice(4, p=pi)
  ################################
  return action
```

```python
def softmax(theta):
  tot = np.zeros([len(theta)])
  exp = np.exp(theta)
  for i in range(len(theta)):
    tot[i] = exp[i]/np.sum(exp)
  return tot
```

```python
# calculate gradient of ln theta 10%
def gradient_ln( action, policy):
  ###### Your code here ######
  gra_pi = np.zeros([len(policy)])
  output = np.zeros([len(policy)])
  for x in range(len(policy)):
    if x==action:
      gra_pi[x] = policy[x]*(1 - policy[action])
    else:
      gra_pi[x] = -policy[x]*policy[action]
    output[x] = gra_pi[x] / policy[x] 
  ##########
  return output
```

```python
# Actor critic 5%
def AC():
    ###### Your code here ######
  h_table = np.zeros([6, 6, 4]) #h_table是theta
  v_table = np.zeros([6, 6])
  pi = np.ones([6, 6, 4]) * 0.25  #策略
  returns = []
  for episode in range(500):
    state=S0
    total_reward=0
    while state != ST:
      action = chooseAction_pg(pi[state[0]][state[1]])
      next_state,reward = step1(state,ACTIONS[action])
      delta = reward + GAMMA * v_table[next_state[0]][next_state[1]] - v_table[state[0]][state[1]]
      h_table[state[0]][state[1]] += ALPHA_theta * delta * gradient_ln(action, pi[state[0]][state[1]]) ##更新policy
      pi[state[0]][state[1]] = softmax(h_table[state[0],state[1]]) ##更新pi的機率
      v_table[state[0]][state[1]] += ALPHA_v * delta           
      state=next_state
      total_reward += reward
    returns.append(total_reward)    
    ################################
  return returns
```

```python
test=moving_averages(np.average([AC() for i in range(5)],axis=0))
```

```python
class ShortCorridor:
    """
    Short corridor environment
    """
    def __init__(self):
        self.reset()

    def reset(self):
        self.state = 0

    def step(self, go_right):
        o1=True
        o2=0
        ###### Your code here ######
        self.go_right = go_right
        if self.state == 0:
          if self.go_right == 1:
            self.state = self.state + 1
          else:
            self.state = self.state
        else:
          if self.go_right == 1:
            self.state = self.state - 1
          else:
            self.state = self.state + 1

            
        if self.state == 3:
          self.reward = 0
          o1 = True
        else:
          self.reward = -1
          o1 = False
        o2 = self.reward    
        ################################
        return [o1, o2]
```

```python
EPSILON=0.1
ALPHA=0.2
GAMMA=1.0
S0 = 0
ST = 3
ACTIONS = [0,1]
def chooseAction(state, qtable):
    if rd.rand()>=EPSILON:
        return np.argmax(qtable[state])
    else:
        return rd.randint(2)

def Q_learning():
  ###### Your code here ######
  env = ShortCorridor()
  qtable = np.zeros([4,2])
  returns = []
  for episode in range(500):
    env.reset()
    #print(env.state)
    total_reward = 0
    che_state = []
    while env.state != ST:
      action = chooseAction(env.state, qtable)
      state = env.state
      env.step(ACTIONS[action])
      state_n = env.state
      reward = env.reward
      che_state.append(env.state)
      qtable[state, action] += ALPHA * (reward + GAMMA * np.max(qtable[state_n]) - qtable[state,action])
      total_reward += reward
          
      #print(total_reward)
    returns.append(total_reward)
  ################################
  return returns
```

### 2.4(15%)Actor-Critic
Please apply Actor-Critic in the environment with $\gamma=1 $,  learning rate for policy $\alpha_\theta=0.001$, learning rate for value state $\alpha_v=0.01$ over 500 episodes.
 The output 'returns' is a list consisting of return 'G' from each episode.



```python
GAMMA = 1.0
ALPHA_theta = 0.001
ALPHA_v = 0.01

# Actor critic
def chooseAction_pg(pi):
    ###### Your code here ######
  action = np.random.choice(2, p=pi)
    ################################
  return action

# Actor critic
def gradient_ln( action, policy):
  ###### Your code here ######
  gra_pi = np.zeros([len(policy)])
  output = np.zeros([len(policy)])
  for x in range(len(policy)):
    if x==action:
      gra_pi[x] = policy[x]*(1 - policy[action])
    else:
      gra_pi[x] = -policy[x]*policy[action]
    output[x] = gra_pi[x] / policy[x] 
  ################################
  return gra_pi

def softmax(theta):
  tot = np.zeros([len(theta)])
  exp = np.exp(theta)
  for i in range(len(theta)):
    tot[i] = exp[i]/np.sum(exp)
  return tot

# Actor critic
def AC():
    ###### Your code here ######
  env = ShortCorridor()
  theta_table = np.zeros([4,2]) #h_table是theta
  v_table = np.zeros([4])
  pi = np.ones([4,2]) * 0.5  
  returns = []
  for episode in range(500):
    state=S0
    env.reset()
    G=0
    while state != ST:
      action=chooseAction_pg(pi[state])
      env.step(ACTIONS[action])
      state_n = env.state
      reward = env.reward
      delta=reward + GAMMA*v_table[state_n] - v_table[state]
      grad_ln=gradient_ln(action, pi[state]) 
      theta_table[state] += ALPHA_theta*delta*grad_ln ##更新policy
      pi[state] = softmax(theta_table[state])##更新pi的機率
      v_table[state] += ALPHA_v*delta           
      state=state_n
      G+=reward
    returns.append(G)       
    ################################
  return returns
```
